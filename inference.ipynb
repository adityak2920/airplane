```python
trans = tf.Compose([tf.Resize((224, 224)),
                    tf.ToTensor(),
                    tf.Normalize(mean=(0.485, 0.456, 0.406), std=(0.229, 0.224, 0.225)),
                    ])
```

```python
# model = torch.load("pure_model.pth", map_location=torch.device('cpu'))
model = torch.load("pure_pyfast.pth", map_location=torch.device('cpu'))
for i in os.listdir("/Users/adityakumar/Documents/pulse/val/Embraer ERJ 145/"):
    image = Image.open("/Users/adityakumar/Documents/pulse/val/Embraer ERJ 145/"+i)
    trans_img = trans(image)
    pred = model(trans_img.unsqueeze(0))
    print(classes[pred.cpu().detach().numpy().argmax(1)[0]])
    
```
